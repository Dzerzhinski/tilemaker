```python
FILENAME = "testfile" 
dX = 2000 
X_OFFSET = 1000
dY = 1732
Y_OFFSET = 866

X_MIDEDGE_1 = 250
X_MIDEDGE_2 = 1000
X_MIDEDGE_3 = 1750

Y_MIDEDGE_1 = 0
Y_MIDEDGE_2 = 433
Y_MIDEDGE_3 = 1299
Y_MIDEDGE_4 = 1732

STROKEWIDTH = 100 

T0 = [] 
T1 = [[X_MIDEDGE_3, Y_MIDEDGE_2, X_MIDEDGE_3, Y_MIDEDGE_3]]
T2 = [[X_MIDEDGE_3, Y_MIDEDGE_2, X_MIDEDGE_2, Y_MIDEDGE_4]]
T3 = [[X_MIDEDGE_2, Y_MIDEDGE_1, X_MIDEDGE_2, Y_MIDEDGE_4]] 
T4 = T1 + [X_MIDEDGE_1, Y_MIDEDGE_3, X_MIDEDGE_2, Y_MIDEDGE_4] 
T5 = T1 + [X_MIDEDGE_1, Y_MIDEDGE_2, X_MIDEDGE_2, Y_MIDEDGE_3] 
T6 = T1 + T3[0] 

def arc_dfn(file): 
    file.write("<path id=\"c-arc\" d=\"M1750 433 Q1000 866, 1750 1299\" fill=\"transparent\" />\n")
    file.write("<path id=\"cat-arc\" d=\"M1750 433 Q1000 866, 1000 1732\" fill=\"transparent\" />\n")
    file.write("<path id=\"straight-arc\" d=\"M1000 0, 1000 1732\" fill=\"transparent\" />\n") 
    file.write("<path id=\"triple-arc\" d=\"M1000 0 Q2000 866, 1000 1732\" fill=\"transparent\" />\n\n")
    
def mask_dfn(file): 
    # T2R1 
    file.write("<mask id=\"MT2R1\" >\n") 
    file.write("\t<rect height=\"100%\" width=\"100%\" fill=\"white\" />\n") 
    file.write("\t<use xlink:href=\"#cat-arc\" transform=\"rotate(60, 1000, 866)\" stroke=\"black\" ") 
    file.write("stroke-width=\"{}\" />\n".format(2 * STROKEWIDTH))
    file.write("</mask>\n\n")
    
def tile_dfn_1(file): 
    # T1
    file.write("<g id=\"T1\" >\n")
    file.write("\t<use xlink:href=\"#c-arc\" stroke=\"blue\" stroke-width=\"100\" />\n")
    file.write("</g>\n\n")
    # T2 
    file.write("<g id=\"T2\" >\n") 
    file.write("\t<use xlink:href=\"#cat-arc\" stroke=\"blue\" stroke-width=\"100\" />\n") 
    file.write("</g>\n\n") 
    # T3 
    file.write("<g id=\"T3\" >\n") 
    file.write("\t<use xlink:href=\"#straight-arc\" stroke=\"blue\" stroke-width=\"100\" />\n") 
    file.write("</g>\n\n") 
    
def tile_dfn_2(file):
    # T4
    file.write("<g id=\"T4\" >\n") 
    file.write("\t<use xlink:href=\"#T1\" />\n") 
    file.write("\t<use xlink:href=\"#T1\" transform=\"rotate(120, 1000, 866)\" />\n") 
    file.write("</g>\n\n")
    # T5 
    file.write("<g id=\"T5\" >\n") 
    file.write("\t<use xlink:href=\"#T1\" />\n") 
    file.write("\t<use xlink:href=\"#T1\" transform=\"rotate(180, 1000, 866)\" />\n") 
    file.write("</g>\n\n") 
    # T6 
    file.write("<g id=\"T6\" >\n") 
    file.write("\t<use xlink:href=\"#T1\" />\n") 
    file.write("\t<use xlink:href=\"#T3\" />\n") 
    file.write("</g>\n\n") 
    # T7 
    file.write("<g id=\"T7\" >\n") 
    file.write("\t<use xlink:href=\"#T1\" />\n") 
    file.write("\t<use xlink:href=\"#T2\" transform=\"rotate(120, 1000, 866)\" />") 
    file.write("</g>\n\n") 
    # T8 
    file.write("<g id=\"T8\" >\n") 
    file.write("\t<use xlink:href=\"#T2\" />\n") 
    file.write("\t<use xlink:href=\"#T2\" transform=\"rotate(120, 1000, 866)\" />") 
    file.write("</g>\n\n") 
    # T9 
    file.write("<g id=\"T9\" >\n") 
    file.write("\t<use xlink:href=\"#T2\" transform=\"rotate(60, 1000, 866)\" />\n") 
    file.write("\t<use xlink:href=\"#T3\" mask=\"url(#MT2R1)\" />\n")
    file.write("</g>\n\n")
    
def tx_dfn(file): 
    file.write("<g id=\"T1R1\" >\n")
    file.write("\t<use xlink:href=\"#T1\" transform=\"rotate(60, 1000, 866)\" />\n") 
    file.write("</g>\n\n")
      

def draw_tile(file, x_pos = 0, y_pos = 0, stroke = "black", stroke_width = 30, 
              stroke_color = "black", scale = 0.10, fill = "transparent", 
              fill_opacity = 1.0): 
    file.write("<use xlink:href=\"#hextile\" ")
    file.write("x=\"{}\" ".format(int(x_pos)))
    file.write("y=\"{}\" ".format(int(y_pos))) 
    file.write("stroke=\"{}\" ".format(stroke)) 
    file.write("stroke-width=\"{}\" ".format(int(stroke_width))) 
    file.write("stroke-color=\"{}\" ".format(stroke_color)) 
    file.write("transform=\"scale({:.2f})\" ".format(scale)) 
    file.write("fill=\"{}\" ".format(fill)) 
    file.write("fill-opacity=\"{:.2f}\" ".format(fill_opacity)) 
    file.write(" />\n")
     
    
def draw_corner_arc(file):  
    file.write("<path d=\"M1000 0 Q1250 433, 1500 433\" ") 
    file.write("fill=\"transparent\" ") 
    file.write("/>\n") 
    
    
def corona_coords(kappa, theta, x_ht, y_ht, origin_x, origin_y): 
    dx = origin_x
    dy = origin_y
    if(kappa == 0): 
        return dx, dy
    y_off = int(y_ht / 2)
    x_off = int((3 * x_ht) / 4)
    sextant = theta // kappa 
    s = theta % kappa
    if(sextant == 0): 
        dy = origin_y - (y_ht * (kappa)) + (s * y_off) 
        dx = origin_x + (s * x_off)
    elif(sextant == 1): 
        dy = origin_y - (y_off * (kappa)) + (s * y_ht) 
        dx = origin_x + ((kappa) * x_off)
    elif(sextant == 2): 
        dy = origin_y + (y_off * (kappa)) + (s * y_off) 
        dx = origin_x + ((kappa - s) * x_off)
    elif(sextant == 3): 
        dy = origin_y + (y_ht * (kappa)) - (s * y_off) 
        dx = origin_x - (s * x_off)
    elif(sextant == 4): 
        dy = origin_y + (y_off * (kappa)) - (s * dY) 
        dx = origin_x - ((kappa) * x_off)
    elif(sextant == 5): 
        dy = origin_y - (y_off * (kappa)) - (s * y_off) 
        dx = origin_x - ((k - s + 1) * x_off)
    return dx, dy


svgfile = open(FILENAME + ".svg", "w") 

svgheader = "<?xml version=\"1.0\"?>\n" +\
            "<svg xmlns=\"http://www.w3.org/2000/svg\" xmlns:xlink=\"http://www.w3.org/1999/xlink\" >" +\
            "\n" +\
            "<defs>\n" +\
            "\t<polygon id=\"hextile\" stroke=\"black\" stroke-width=\"30\" " +\
            "points=\"500,0 1500,0 2000,866 1500,1732 500,1732 0,866\" />\n" +\
            "\n"

svgfile.write(svgheader) 
arc_dfn(svgfile) 
tile_dfn_1(svgfile) 
mask_dfn(svgfile)
tile_dfn_2(svgfile)
tx_dfn(svgfile)
svgfile.write("</defs>\n\n")

#svgfile.write("<use xlink:href=\"#hextile\" fill=\"white\" />\n")
#for i in range(5): 
#    for j in range(5):
#        svgfile.write("<use xlink:href=\"#hextile\" x = \"" + str(j * (dX + X_OFFSET) + (i % 2 * 1500)) +\
#                      "\" y = \"" + str(i * Y_OFFSET) + "\" fill=\"white\" transform=\"scale(0.1)\" />\n")
        
#CORONA = 12
#for k in range(CORONA + 1): 
#    dx1 = (CORONA - k) * 1500 
#    dx2 = (CORONA + k) * 1500 #

#    for j in range(2 * CORONA + 1 - k): 
#        dy = (k * Y_OFFSET) + (j * dY)
#        svgfile.write("<use xlink:href=\"#hextile\" x=\"" + str(dx1) + "\" y=\"" + str(dy) +\
#                      "\" fill=\"white\" transform=\"scale(0.1)\" />\n") 
#        svgfile.write("<use xlink:href=\"#hextile\" x=\"" + str(dx2) + "\" y=\"" + str(dy) +\
#                      "\" fill=\"white\" transform=\"scale(0.1)\" />\n") 

CORONA = 5 
ORIGIN_X = (CORONA + 2) * 1250 
ORIGIN_Y = (CORONA + 1) * dY

svgfile.write("<use xlink:href=\"#hextile\" x=\"" + str(ORIGIN_X) + "\" y=\"" + str(ORIGIN_Y) +\
              "\" fill=\"white\" transform=\"scale(0.1)\" />\n")

for k in range(CORONA):
    opacity = (k + 1) / (2 * CORONA)
    for t in range (6 * (k + 1)): 
        x_pos, y_pos = corona_coords(k + 1, t, dX, dY, ORIGIN_X, ORIGIN_Y)
        draw_tile(svgfile, x_pos, y_pos, fill = "blue", fill_opacity = opacity)
        #print("Tile: ({}, {})".format(k + 1, t))
    
#svgfile.write("<use xlink:href=\"#T9\" x=\"{}\" y=\"{}\" transform=\"scale(0.1)\" />\n\n".format(ORIGIN_X, ORIGIN_Y))
        
svgfile.write("</svg>\n") 
svgfile.close()


            
```
